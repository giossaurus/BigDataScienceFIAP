```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/Gradua%C3%A7%C3%A3o/dados/diamonds.csv")
```

```python
df
```

```python
#EstatÃ­sticas descritivas da populaÃ§Ã£o
# MÃ©dia aritmÃ©tica
df.price.mean()
```

```python
#Outro jeito de escrever vÃ¡lido tambem
df[['price']].mean()
```

```python
#Mediana
df[['price']].median()
```

```python
#Desvio PadrÃ£o
df[['price']].std()
```

```python
# GeraÃ§Ã£o de uma amostra e ver se representa a populaÃ§Ã£o
a1 = df.iloc[0:3000, :]
print(a1[['price']].mean())
print(a1[['price']].median())
print(a1[['price']].std())
```

```python
# GeraÃ§Ã£o de uma segunda amostra e ver se representa a populaÃ§Ã£o
a2 = df.iloc[3000:6000, :]

print(a2[['price']].mean())
print(a2[['price']].median())
print(a2[['price']].std())
```

```python
# Como as estatÃ­sticas descritivas sÃ£o muito diferentes (>5%), concluimos que a1 e a2 nÃ£o representam a populaÃ§Ã£o
```

```python
# GeraÃ§Ã£o de amostra aleatÃ³ria
a3 = df.sample(3000, random_state=1)
print(a3[['price']].mean())
print(a3[['price']].median())
print(a3[['price']].std())
```
