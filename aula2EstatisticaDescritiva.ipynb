```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/Gradua%C3%A7%C3%A3o/dados/diamonds.csv")
```

```python
df
```

```python
#Estatísticas descritivas da população
# Média aritmética
df.price.mean()
```

```python
#Outro jeito de escrever válido tambem
df[['price']].mean()
```

```python
#Mediana
df[['price']].median()
```

```python
#Desvio Padrão
df[['price']].std()
```

```python
# Geração de uma amostra e ver se representa a população
a1 = df.iloc[0:3000, :]
print(a1[['price']].mean())
print(a1[['price']].median())
print(a1[['price']].std())
```

```python
a1.shape
```

```python
a1.head()
```

```python
# Geração de uma segunda amostra e ver se representa a população
a2 = df.iloc[3000:6000, :]

print(a2[['price']].mean())
print(a2[['price']].median())
print(a2[['price']].std())
```

```python
a2.shape
```

```python
a2.head()
```

```python
# Como as estatísticas descritivas são muito diferentes (>5%), concluimos que a1 e a2 não representam a população
```

```python
# Geração de amostra aleatória
a3 = df.sample(3000, random_state=1)
print(a3[['price']].mean())
print(a3[['price']].median())
print(a3[['price']].std())
```
