```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP-On-Line-Graduation/main/datasets/%20mtcars.csv")
```

```python
df
```

```python
# Análise exploratória
df.plot.scatter(x="wt", y="mpg")
```

```python
# KMeans
k = KMeans(n_clusters=7, random_state=1)
k.fit(df[["wt", "mpg"]])
```

```python
# 4 clusters
df['cluster'] = k.labels_
df.plot.scatter(x="wt", y="mpg", c="cluster", colormap="rainbow")
```
